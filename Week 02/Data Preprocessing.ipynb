## 2.2.  Statistics - Preprocessing

Mathematical Formulation with simple example: 
    
1. Mean 
($\bar{X}$) : $\sum_{i=1}^{n} \frac{x_i}{n}$ = $\frac{4 + 8 + 12 + 21 + 33 + 58 + 92 + 98}{8} = 40.75$



2. Variance (V) : $\sum_{i=1}^{n} \frac{(x_i -\bar{x})^2 }{n}$ = $\frac{ (4 - 40.75)^2  +  (8-40.75)^2 + (12-40.75)^2 + (21-40.75)^2 + (33-40.75)^2 + (58-40.75)^2 + (92-40.75)^2 + (98-40.75)^2}{8} = 1237.7$ 



3. Standard Deviation (σ) : $\sqrt{\sum_{i=1}^{n} \frac{(x_i -\bar{x})^2 }{n}}$ = $\sqrt{Variance(V)}$ = $\sqrt{1237.7} = 35.2 $



4. Median (M) = $\frac{21 + 33}{2} = 27 $

#### 2.5.6.4. Impute missing values with Mean, Median, and Mode values

A question may arise in our mind, which is better replacement by mean, replacement by median, or replacement by mode? 


Answer: It always depends on our data and on our tasks. Often a simple, if not always satisfactory, the choice for missing values that are known not to be zero is to use some ``central'' value of the variable. This is often the mean, median, or mode, and thus usually has limited impact on the distribution. We might choose to use the mean, for example, if the variable is otherwise generally normally distributed (and in particular does not have any skewness). If the data does exhibit some skewness though (e.g., there are a small number of very large values) then the median might be a better choice.

For categoric variables, there is, of course, no mean nor median, and so in such cases, we might choose to use the mode (the most frequent value) as the default to fill in for the otherwise missing values. The mode can also be used for numeric variables.

### 2.6.1. How to calculate outlier Interquartile Range Rule

The interquartile range rule is useful in detecting the presence of outliers. Outliers are individual values that fall outside of the overall pattern of a data set. This definition is somewhat vague and subjective, so it is helpful to have the rule to apply when determining whether a data point is truly an outlier—this is where the interquartile range rule comes in.

#### 2.6.1.1. What Is the Interquartile Range?

Any set of data can be described by its five-number summary. These five numbers, which give you the information you need to find patterns and outliers, consist of (in ascending order):

1. The median of the data set, which represents the midpoint of the whole list of data

2. The first quartile Q1, which represents a quarter of the way through the list of all data

3. The third quartile Q3, which represents three-quarters of the way through the list of all data

4. The minimum or lowest value of the dataset

5. The maximum or highest value of the data set.

Let's take an example: 

Data = [4, 2, 5, 3, 7, 5, 6, 9, 32, 2]

Arranging Data in Ascending order: [2, 2, 3, 4, 5, 5, 6, 7, 9, 32]

Median = $\frac{5+5}{2} = 5$

After Calculating median, our new data = [2, 2, 3, 4, 5, 5, 5, 6, 7, 9, 32]

The first quartile Q1 = $\frac{ 6+ 7}{2} = 3.5$

The third quartile Q3 = $\frac{6 + 7}{2} = 6.5$

Minimum or lowest value of the data = 2 

Maximum or highest value of the data = 9 

The interquartile range is calculated in much the same way as the range. All we do to find it is subtract the first quartile from the third quartile:

interquartile Range Formula = ${Q_3} - {Q_1} = 6.5 - 3.5 = 3$

There are two standard to find outlier:

1. Mild Outlier: A point beyond an inner fence on either side is considered a mild outlier.

Q1 - 1.5 * IQR(Lower) = 3.5 - 1.5 * 3 = -1.5 <br>
Q3 + 1.5 + IQR(Upper) = 6.5 + 1.5 * 3 = 11  

2. Extreme Outlier: A point beyond an outer fence is considered an extreme outlier. 

Q1 - 3 * IQR(Lower) = 3.5 - 3 * 3 = -5.5 <br>
Q3 + 3 + IQR(Upper) = 6.5 + 3 * 3 = 15.5  <br>

Note: The outlier is identified as the largest value in the data set, 32, and appears as the circle to the right of the box plot.

## 2.8. Dummy Coding - Preprocessing  

Dummy variables (also known as one-hot encoding) consists on transforming a categorical feature into a form that is better understood by several algorithms. Let’s say we have a feature called color which has one of three options (red, green and blue). Most algorithms can’t interpret the text so we would need to convert each value into a number. We could potentially replace red with 1, green with 2 and blue with 3. This could work, but it has an underlying assumption that red is closer to green than to blue. It can also be interpreted as blue being more important than red since it has a higher value. This may not necessarily be the case. A solution that addresses these challenges is to apply dummy variables. We can replace the color feature with three new features called red, green and blue. This would be the breakdown for each color:

|Color      |      Red          |   Green     |   Green     | 
|:--------  |:-----------------:|:------------|:------------|
|Red        | 1                 | 0           |      0      |
|Green      | 0                 | 1           |      0      |
|Blue       | 1                 | 0           |      1      |

In this case, we no longer indicate that one color is more important than any other, or that two colors are more similar than others. Note that we could also get rid of one of the columns (i.e. the Blue column), and assume that if both Red and Green have 0’s, then this automatically indicates that Blue is equal to 1.

In our last observation the variable that should take values as 0 and 1, where the values indicate the presence or absence peak. 

Example: A 0 may indicate a Peak and 1 may indicate an offpeak). Where a categorical variable has more than two categories, it can be represented by a set of dummy variables, with one variable for each category. 


Note: Numeric variables can also be dummy coded to explore nonlinear effects. Dummy variables are also known as indicator variables, design variables, contrasts, one-hot coding, and binary basis variables

## 2.9. Normalization - Preprocessing 

When we construct a building, house, or road, there is a leveling done before the construction starts. Everything is brought to a level and then the building is erected.

Likewise, normalization is done so that all the data is at the same level and then we can run whatever algorithm is used. Building these data-related solutions requires the data to be brought to a level before any algorithm is applied. 

So, Normalization is one of the methods used in data science to bring features in a dataset to the same scale. When we normalize a feature, all feature values will be in the range of 0 to 1.

In python, we have several packages for Normalization. 

1. Min Max Scaler Sub-Package: $\frac{x-x_{min}}{x_{max} - x_{min}}$

2. Normalize Sub Package: <br>
    a. L1 Norm (Manhattan Distance): $ \frac{x_{ij}}{|x_1| + |x_2| + |x_3| + .... +|x_n|} $ 
   
    b. L2 Norm (Euclidean Distance Distance): $ \frac{x_{ij}}{\sqrt{{x_1}^2 + {x_2}^2 + {x_3}^3 + .... +{x_n}^2}} $ 

### 2.9.1. Normalize our preprocessed data 

```python
from sklearn.preprocessing import minmax_scale, normalize
```

```python
# First Method: Min Max Scale

data_set_11 = minmax_scale(data_set_10.iloc[:,0:5], feature_range=(0,1))

# Second Method: using L2 Normalization

data_set_12 = normalize(data_set_10.iloc[:,0:5], norm='l2', axis = 0)

# axis = 0 for normalizing features where axis = 1 is for normalizing each sample
```
